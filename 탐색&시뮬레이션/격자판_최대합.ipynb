▣ 출력예제 1

```python
# 내 풀이
n = int(input())
arr=[]
max = 0

# 2차원 배열 새성
for _ in range(n):
    a = list(map(int,input().split()))
    arr.append(a)

# 가로에서 가장 큰값
for x in range(n):
    new_max=0
    for y in range(n):
        new_max += arr[x][y]
    if new_max >max:
        max = new_max
# 세로에서 가장 큰값
for y in range(n):
    new_max=0
    for x in range(n):
        new_max += arr[x][y]
    if new_max >max:
        max = new_max
# 대각선
x=0
y=0
new_max=0
for _ in range(5):
    new_max += arr[x][y]
    x+=1
    y+=1
    if new_max >max:
        max = new_max

# 반대 대각선
x=0
y=4
new_max=0
for _ in range(5):
    new_max += arr[x][y]
    x+=1
    y-=1
    if new_max >max:
        max = new_max
print(max)
```

```python
# 강의 풀이

n = int(input())
a = [list(map(int,input().split())) for _ in range(n)]
largest=-1111
for i in range(n):
    sum1 = sum2 = 0 # sum1 = 행의합, sum2 = 열의합
    for j in range(n):
        sum1 += a[i][j]
        sum2 += a[j][i]
        if sum1>largest:
            largest = sum1
        if sum2 >largest:
            largest = sum2
sum1 = sum2 = 0 # 대각선을 위해 초기화
for i in range(n):
    sum1 += a[i][i] # 왼쪽에서 오른쪽으로 가는 대각선
    sum2 += a[i][n-i-1] # 오른쪽에서 왼쪽으로 가는 대각선
    if sum1>largest:
            largest = sum1
    if sum2 >largest:
        largest = sum2

print(largest)
```
