```python
# 내풀이
# 잘 뒤집기만 할줄 알면 되지 않을까..?

# [1~20] 배열 만들기
arr = [x for x in range(1,21)]

# 10번 입력을 받고 입력을 받으면 그 인덱스 안에 배열만큼 삭제하고 다시 추가한다.
for _ in range(10):
    start, end = map(int,input().split())
    new_arr = arr[start-1:end]
    del arr[start-1:end]
    for x in new_arr:
        arr.insert(start-1,x)
        
for x in arr:
    print(x, end=' ')
```

```python
# 강의 풀이

a = list(range(21))
for _ in range(10):
    s, e = map(int,input().split())
    # 여기서 바꿔주는 작업
    for i in range((e-s+1)//2):
        a[s+i],a[e-i] =  a[e-i],a[s+i]
# 이제 출력할때 a[0]만 버리면된다 
a.pop(0)
for x in a:
    print(x, end=' ')
```
